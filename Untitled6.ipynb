```python
# --- Imports ---
import os
import json
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from tqdm import tqdm
from collections import deque
import gym
from gym import spaces
```

```python
# --- Upload files if not present ---
json_path = "/content/extracted_data.json"
pth_path = "/content/NavigatedLearningDQN.pth"
```

```python
# --- Load coordinates from extracted_data.json ---
with open("extracted_data.json", "r") as f:
    extracted_data = json.load(f)
```

```python
SCALE = 100
resource_cells = []
for name, data in extracted_data.items():
    x = int(float(data["x_coordinate"]) * SCALE)
    y = int(float(data["y_coordinate"]) * SCALE)
    resource_cells.append((x, y))

max_coord = max(max(x, y) for x, y in resource_cells)
GRID_SIZE = max_coord + 1

# --- Define Grid Environment ---
class GridEnv(gym.Env):
    def __init__(self, resources, grid_size=GRID_SIZE):
        super(GridEnv, self).__init__()
        self.grid_size = grid_size
        self.start_pos = (grid_size - 1, 0)
        self.goal_pos = (0, grid_size - 1)
        self.agent_pos = self.start_pos

        self.action_space = spaces.Discrete(2)  # 0: UP, 1: RIGHT
        self.observation_space = spaces.Discrete(grid_size * grid_size)
        self.resources = set(resources)

    def reset(self):
        self.agent_pos = self.start_pos
        return self._get_state(), {}

    def step(self, action):
      if action == 0 and self.agent_pos[0] > 0:  # Up
          self.agent_pos = (self.agent_pos[0] - 1, self.agent_pos[1])
      elif action == 1 and self.agent_pos[1] < self.grid_size - 1:  # Right
          self.agent_pos = (self.agent_pos[0], self.agent_pos[1] + 1)

      reward = 1.0 if self.agent_pos in self.resources else -0.1
      done = False

      if self.agent_pos == self.goal_pos:
          reward += 10.0  # ðŸŽ¯ Reward for reaching the goal
          done = True

      next_state = self._get_state()
      info = {}

      return next_state, reward, done, info



    def _get_state(self):
        return self.agent_pos[0] * self.grid_size + self.agent_pos[1]

# --- Define DQN Model ---
class DQN(nn.Module):
    def __init__(self, state_size, action_size=2):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(state_size, 128)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, action_size)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        return self.fc2(x)

# --- Set up environment and model ---
env = GridEnv(resources=resource_cells, grid_size=GRID_SIZE)
state_size = env.observation_space.n
action_size = 2  # âœ… Match the trained model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- Load trained model ---
model = DQN(state_size, action_size).to(device)
model.load_state_dict(torch.load("NavigatedLearningDQN.pth", map_location=device))
model.eval()

# --- Simulate until agent reaches goal (or max_steps to avoid infinite loop) ---
state, _ = env.reset()
done = False
path = [env.agent_pos]

print("ðŸ” Simulating agent path...\n")
steps = 0
max_steps = 10000  # safety cap to prevent infinite loops

pbar = tqdm(total=max_steps, desc="Agent Moving")
while not done and steps < max_steps:
    with torch.no_grad():
        state_tensor = torch.eye(state_size)[state].unsqueeze(0).to(device)
        q_values = model(state_tensor)
        action = torch.argmax(q_values).item()
    result = env.step(action)
    if result is None:
        print("âš ï¸ step() returned None â€” make sure the environment is working.")
        break
    next_state, _, done, _ = result
    state = next_state
    path.append(env.agent_pos)
    steps += 1
    pbar.update(1)
pbar.close()


# --- Visualization function ---
def render_path(path, resources, start, goal, grid_size):
    grid = np.zeros((grid_size, grid_size))

    for x, y in resources:
        grid[x, y] = 0.6  # ðŸ“– Resources

    for x, y in path:
        grid[x, y] = 0.3  # ðŸŸ¢ Agent path

    sx, sy = start
    gx, gy = goal
    grid[sx, sy] = 0.1  # ðŸ”µ Start
    grid[gx, gy] = 0.9  # ðŸ”´ Goal

    plt.figure(figsize=(7, 7))
    plt.imshow(grid, cmap="Spectral", origin="upper")
    plt.title("DQN Agent Simulation")
    plt.xticks(np.arange(0, grid_size, max(1, grid_size // 10)))
    plt.yticks(np.arange(0, grid_size, max(1, grid_size // 10)))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.colorbar(label="Cell Types")
    plt.show()

# --- Show Final Simulation ---
render_path(path, resource_cells, env.start_pos, env.goal_pos, env.grid_size)
```
